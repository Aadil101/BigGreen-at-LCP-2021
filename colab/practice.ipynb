```
# load task info
task = 'lcp'#'momi'
task_defs = TaskDefs('data_complex/lcp.yml')#TaskDefs('data_momi/momi.yml')
assert task in task_defs._task_type_map
assert task in task_defs._data_type_map
assert task in task_defs._metric_meta_map
prefix = task.split('_')[0]
task_def = task_defs.get_task_def(prefix)
data_type = task_defs._data_type_map[task]
task_type = task_defs._task_type_map[task]
metric_meta = task_defs._metric_meta_map[task]
# load model
checkpoint_path = 'checkpoints/bert-cased_lcp-single_2020-10-24T2321/model_4.pt'#'checkpoints/bert-cased_momi_2020-11-05T0130/model_1.pt'#'checkpoints/bert-cased_lcp-single_2020-10-29T1723/model_4.pt'
assert os.path.exists(checkpoint_path)
state_dict = torch.load(checkpoint_path)
config = state_dict['config']
config["cuda"] = True
task_def = task_defs.get_task_def(prefix)
task_def_list = [task_def]
config['task_def_list'] = task_def_list
## temp fix
config['fp16'] = False
config['answer_opt'] = 0
config['adv_train'] = False
del state_dict['optimizer']
config['output_attentions'] = True
model = MTDNNModel(config, state_dict=state_dict)
encoder_type = config.get('encoder_type', EncoderModelType.BERT)
```

```
do_lower_case = False
root = 'data_complex/'#'data_momi/'
literal_model_type = 'bert-base-cased'.split('-')[0].upper()
encoder_model = EncoderModelType[literal_model_type]
literal_model_type = literal_model_type.lower()
mt_dnn_suffix = literal_model_type
if 'base' in 'bert-base-cased':
    mt_dnn_suffix += "_base"
elif 'large' in 'bert-base-cased':
    mt_dnn_suffix += "_large"
```

```
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
from tqdm import tqdm
import torch
import pandas as pd
import pickle5 as pkl
device = 'cuda'
model_id = 'gpt2-large'
model = GPT2LMHeadModel.from_pretrained(model_id).to(device)
tokenizer = GPT2TokenizerFast.from_pretrained(model_id)
```

```
from nlp import load_dataset
test = load_dataset('wikitext', 'wikitext-2-raw-v1', split='test')
encodings = tokenizer('\n\n'.join(test['text']), return_tensors='pt')
```

```
encodings['input_ids']
```

```
max_length = model.config.n_positions
stride = 512

lls = []
for i in tqdm(range(0, encodings.input_ids.size(1), stride)):
    begin_loc = max(i + stride - max_length, 0)
    end_loc = i + stride
    input_ids = encodings.input_ids[:,begin_loc:end_loc].to(device)
    target_ids = input_ids.clone()
    target_ids[:,:-stride] = -100

    with torch.no_grad():
        outputs = model(input_ids, labels=target_ids)
        log_likelihood = outputs[0] * stride

    lls.append(log_likelihood)

ppl = torch.exp(torch.stack(lls).sum() / i)
```
